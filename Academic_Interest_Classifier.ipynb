Data Cleaning

```python
import re
# Preprocessing steps
def clean_text(text):
    # Remove punctuation (titik, koma, dan tanda baca lainnya)
    text = re.sub(r'[^\w\s]', '', text)
    # Remove extra white spaces
    text = re.sub(r'\s+', ' ', text)
    # Convert text to lowercase
    text = text.lower()
    return text

# Clean 'Sub_bidang' column
data['Sub_bidang'] = data['Sub_bidang'].apply(clean_text)

# Clean 'Prodi_Kuliah' column
data['Prodi_Kuliah'] = data['Prodi_Kuliah'].apply(clean_text)

# Clean 'Prodi_SLTA' column
data['Prodi_SLTA'] = data['Prodi_SLTA'].apply(clean_text)
```

#### Stopwords Removal

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words("indonesian"))

additional_stopwords = ["dan", "yang", "itu", "dengan"]
stop_words.update(additional_stopwords)

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

motivasi_lemmatized_no_stopwords = motivasi_lemmatized.apply(remove_stopwords)
aktivitas_luang_lemmatized_no_stopwords = aktivitas_luang_lemmatized.apply(remove_stopwords)
tema_bacaan_lemmatized_no_stopwords = tema_bacaan_lemmatized.apply(remove_stopwords)

all_lemmatized_no_stopwords = motivasi_lemmatized_no_stopwords + aktivitas_luang_lemmatized_no_stopwords + tema_bacaan_lemmatized_no_stopwords


print("Hasil Lemmatisasi Tanpa Stopwords:")
print(all_lemmatized_no_stopwords)


max_length = max([len(sentence) for sentence in all_lemmatized_no_stopwords])
print(f"Panjang Maksimum Sequence Setelah Penghapusan Stopwords: {max_length}")
```

```python
from nltk import FreqDist
from nltk.tokenize import word_tokenize


all_tokens = [word for sentence in all_lemmatized_no_stopwords for word in sentence]

freq_dist = FreqDist(all_tokens)
mean_frequency = sum(freq_dist.values()) / len(freq_dist)
threshold = mean_frequency

low_frequency_words = [word for word, freq in freq_dist.items() if freq < threshold]
print("Kata-kata dengan frekuensi rendah:", low_frequency_words)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
from lightgbm import LGBMClassifier
```

```python
data.dropna(subset=['Label_id'], inplace=True)
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Memisahkan fitur dan target
X = data.drop(['Label_id', 'Motivasi', 'Tema_bacaan', 'Aktivitas_Luang'], axis=1)
y = data['Label_id']
```

```python
# Menghapus nilai NaN
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
```

```python
# Oversampling menggunakan SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_imputed, y)
```

```python
import tensorflow as tf
from tensorflow import keras

from keras import layers
from keras import Sequential
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
from tensorflow.keras.utils import plot_model
from keras.layers import Dense, Dropout
```

Experiment with Epoch

```python
def fit_first_model(x_train, y_train, x_test, y_test, epoch):

    # define model
    model = Sequential()
    model.add(layers.Dense(32, input_dim=1400, activation='relu'))
    model.add(layers.Dense(64, activation='relu'))
    model.add(Dense(6, activation='relu'))
    model.add(Dropout(0.2))
    model.add(layers.Dense(1, activation='softmax'))


    # compile model
    model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.CategoricalCrossentropy())

    # fit model
    history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=epoch, verbose=0)

    # plot learning curves
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='test')
    plt.title('epoch='+str(epoch))
    plt.legend()
```

```python
epoch = [50, 100, 150, 200, 250]
```

```python
plt.figure(figsize=(20, 20))
for i in range(len(epoch)):

    # determine the plot number
    plot_no = 420 + (i+1)
    plt.subplot(plot_no)

    # fit model and plot learning curves for a learning rate
    fit_first_model(X_train, y_train, X_test, y_test, epoch[i])

# show learning curves
plt.show()
```
